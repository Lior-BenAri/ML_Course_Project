### According to the article - the target is electrical conductivity. Lets see its statistics:

```python
import pandas as pd

full_dataset_df = pd.read_csv("ossl_all_L0_v1.2.csv")
```

```python
print(f"Number of samples in full dataset: {len(full_dataset_df)}")
visnir_cols = ["scan_visnir.450_ref" , "scan_visnir.2450_ref"]

# filter by Vis-NIR existance:
visnir_dataset_df = full_dataset_df.loc[full_dataset_df.loc[:, visnir_cols[0]:visnir_cols[1]].notna().all(axis=1)]
print(f"Number of samples include Vis-NIR: {len(visnir_dataset_df)}")

USA_elec_cond_col = "ec_usda.a364_ds.m"
Europe_elec_cond_col = "ec_iso.11265_ds.m" 
visnir_USA_EC_dataset_df = visnir_dataset_df[visnir_dataset_df[USA_elec_cond_col].notna()]
visnir_Europe_EC_dataset_df = visnir_dataset_df[visnir_dataset_df[Europe_elec_cond_col].notna()]
print(f"Number of samples include Vis-NIR and USA EC: {len(visnir_USA_EC_dataset_df)}")
print(f"Number of samples include Vis-NIR and Europe EC: {len(visnir_Europe_EC_dataset_df)}")
```

##### Europe EC has 21,782 values. Lets see the histogram 

```python
import matplotlib.pyplot as plt

visnir_Europe_EC_dataset_df[Europe_elec_cond_col].hist(bins=50)
plt.title("Europe EC histogram")
plt.xlabel("EC value")
plt.ylabel("Count")
```

##### From the histogram we that see the variance is too low. Let's check Clay (Soil texture) as target.

```python
# filter by clay existance:
USA_Clay_col = "clay.tot_usda.a334_w.pct"
Europe_Clay_col = "clay.tot_iso.11277_w.pct"
visnir_USA_Clay_dataset_df = visnir_dataset_df[visnir_dataset_df[USA_Clay_col].notna()]
visnir_Europe_Clay_dataset_df = visnir_dataset_df[visnir_dataset_df[Europe_Clay_col].notna()]
print(f"Number of samples include Vis-NIR and USA Clay: {len(visnir_USA_Clay_dataset_df)}")
print(f"Number of samples include Vis-NIR and Europe Clay: {len(visnir_Europe_Clay_dataset_df)}")
```

##### Europe clay has 23,239 values. Lets see the histogram 

```python
import matplotlib.pyplot as plt

visnir_Europe_Clay_dataset_df[Europe_Clay_col].hist(bins=80)
plt.title("Europe EC histogram")
plt.xlabel("Clay value")
plt.ylabel("Count")
```
